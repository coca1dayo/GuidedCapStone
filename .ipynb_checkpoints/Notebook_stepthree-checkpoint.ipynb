```python
# This is part one code for Data Exploration.

# import packages
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Load the Data
filename = 'data_ski.xls'
info = pd.read_excel(filename)

```

```python
# cleaning the NaN with appropiate calculation to make sense of the data
info['fasteight'] = info['fasteight'].fillna(0)
info['terrainparks'] = info['terrainparks'].fillna(0)
info['snow_making_ac'] = info['snow_making_ac'].fillna(0)
info['daysopenlastyear'].fillna((info['daysopenlastyear'].mean()), inplace=True)
info['adultweekday'].fillna((info['adultweekday'].mean()), inplace=True)
info['adultweekend'].fillna((info['adultweekend'].mean()), inplace=True)
info['projecteddaysopen'].fillna((info['projecteddaysopen'].mean()), inplace=True)
info['nightskiing_ac'].fillna((info['nightskiing_ac'].mean()), inplace=True)
```

```python
#checking for completion
info.iloc[140:152]
```

```python
# Checking stats after cleaning data to see how it was affected
info.iloc[140:152].describe()
```

```python
# exploring the data
# looking at the distribution of the price of the adult prices of the weekeday and weekend
x = info['adultweekday']
y = info['adultweekend']
plt.hist(x, bins = 20) # this is the blue graph
plt.xlabel('Frequency')
plt.ylabel('Price')
plt.hist(y, bins = 20) # this is the orange graph
plt.show()
```

```python
# looking at distribution of open days last year with the assumption of the fill in the empty spaces
w = info['daysopenlastyear']
plt.hist(w, bins = 20)
plt.xlabel('Frequency')
plt.ylabel('daysopenlastyear')
plt.show()
```

```python
#looking at the distribution of the average snowfall 
z = info['averageSnowfall']
plt.hist(z, bins = 20)
plt.xlabel('Frequency')
plt.ylabel('Average Snowfall')
plt.show()
```

```python
t = info['projecteddaysopen']
plt.hist(t, bins = 20)
plt.xlabel('Frequency')
plt.ylabel('Projected days open')
plt.show()
```

```python
#Projected days open for the state of Montana
t_m = info['projecteddaysopen'].iloc[140:152]
plt.hist(t_m, bins = 20)
plt.xlabel('Frequency')
plt.ylabel('Projected days open for Montana')
plt.show()
```
